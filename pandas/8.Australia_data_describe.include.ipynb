```python
# параметр include 
# можно указать тип данных, для которого нужно вывести описательную информацию
# для типа данных object метод describe() возвращает DataFrame, в котором указаны:
##	количество непустых строк (count);
##	количество уникальных значений (unique);
##	самое частое значение — мода —  (top);
##	частота — объём использования — этого значения (freq) для каждого столбца типа object исходной таблицы.
```

```python
# для типа данных object 
import pandas as pd
# melb_data.to_csv('data/melb_data.csv', index=False, sep=';')
data = pd.read_csv('data/melb_data.csv', sep = ';')
#display(data)

data.describe(include=['object'])
```
