```python
# Напишите функцию delete_columns(df, col=[]), которая удаляет столбцы из DataFrame и возвращает новую таблицу
import pandas as pd
customer_df = pd.DataFrame({
    'number': [0,1,2,3,4],
    'customer_id':[128, 1201, 9832, 4392, 7472],
    'cust_age': [13, 21, 19, 21, 60],
    'cust_sale': [0, 0, 0.2, 0.15, 0.3],
    'cust_year_birth': [2008, 2000, 2002, 2000, 1961],
    'cust_order': [1400, 14142, 900, 1240, 8430]
})

def delete_columns(df, col=[]):
    for cc in col:
        if cc not in df.columns:
            return None
        return df.drop(col, axis=1)
columns_for_delete = ['number', 'cust_age'] # выбранные вами столбцы
new_df = delete_columns(customer_df, columns_for_delete)
display(new_df)
```

```python
# Рассчитайте среднюю плотность населения представленных стран (количество человек на квадратный километр). 
# Ответ округлите до сотых.
import pandas as pd
countries_df = pd.DataFrame({
    'country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'population': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'square': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})
# Плотность населения рассчитывается как количество человек, проживающих на территории отдельной страны, делённое на площадь этой страны. Обратите внимание, что население в таблице представлено в миллионах
density = countries_df['population']/countries_df['square']*1e6
countries_df['Density'] = density
print('средняя плотность населения', round(countries_df['Density'].mean(),2))
display(countries_df)
# Чтобы рассчитать плотность населения, необходимо столбец с населением (population) умножить на миллион(1e6), чтобы получить общее количество людей, и разделить на столбец с площадью (square)
```
