```python
# ИЗМЕНЕНИЕ ТИПА ДАННЫХ В СТОЛБЦЕ
# воспользоваться методом astype()
# позволяет преобразовать тип данных столбца
# данные в столбцах, которые отражают количество, должны, по идее, выражаться целым числом (Car, Bedroom, Bathroom и Propertyсount), однако кодируются float64 — числом с плавающей запятой размером 64 бита
# при помощи метода astype() последовательно переопределяем столбцы на них же самих, только с изменённым типом данных: int64 — целочисленное число размером 64 бита.
```

```python
import pandas as pd
# melb_data.to_csv('data/melb_data.csv', index=False, sep=';')
data = pd.read_csv('data/melb_data.csv', sep = ';')
#display(data)
data['Car'] = data['Car'].astype('int64')
data['Bedroom'] = data['Bedroom'].astype('int64')
data['Bathroom']= data['Bathroom'].astype('int64')
data['Propertycount'] = data['Propertycount'].astype('int64')
data['YearBuilt'] = data['YearBuilt'].astype('int64')
data.info()
```
