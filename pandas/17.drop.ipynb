```python
# УДАЛЕНИЕ СТОЛБЦОВ
# drop()
# DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
# •	axis — ось совершения операции, axis=0 — удаляются строки, axis=1 — удаляются столбцы
# •	inplace — если параметр выставлен на True, происходит замена изначального DataFrame на новый, при этом метод ничего не возвращает; если на False — возвращается копия DataFrame, из которой удалены указанные строки (столбцы), при этом первоначальный DataFrame не изменяется; по умолчанию параметр равен False.
```

```python
# Удалим столбцы index и Coordinates из таблицы 
import pandas as pd
melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')
#display(melb_data.head())
melb_df = melb_data.copy()
#display(melb_df.head())

melb_df = melb_df.drop(['index', 'Coordinates'], axis=1)
melb_df.head()
```

```python
# Альтернативный вариант:
import pandas as pd
melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')
#display(melb_data.head())
melb_df = melb_data.copy()
#display(melb_df.head())

melb_df.drop(['index','Coordinates'],axis=1,inplace=True)
melb_df.head()
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(12).reshape(3, 4),
                  columns=['A', 'B', 'C', 'D'])
df
```

```python
# Drop columns
df.drop(['B', 'C'], axis=1)
```

```python
# or
df.drop(columns=['B', 'C'])
```

```python
# Drop a row by index
df.drop([0, 1])
```

```python
# Drop columns and/or rows of MultiIndex DataFrame
midx = pd.MultiIndex(levels=[['lama', 'cow', 'falcon'],
                             ['speed', 'weight', 'length']],
                     codes=[[0, 0, 0, 1, 1, 1, 2, 2, 2],
                            [0, 1, 2, 0, 1, 2, 0, 1, 2]])
df = pd.DataFrame(index=midx, columns=['big', 'small'],
                  data=[[45, 30], [200, 100], [1.5, 1], [30, 20],
                        [250, 150], [1.5, 0.8], [320, 250],
                        [1, 0.8], [0.3, 0.2]])
df
```

```python
# Drop a specific index combination from the MultiIndex DataFrame, i.e., drop the combination 'falcon' and 'weight', which deletes only the corresponding row
df.drop(index=('falcon', 'weight'))
```

```python
df.drop(index='cow', columns='small')
```

```python
df.drop(index='length', level=1)
```
